# **Visualization**

```python
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Ville'].value_counts()
```

```python
fig,ax = plt.subplots()
df['Ville'].value_counts().plot(ax = ax, kind = 'bar')
plt.title("The distribution of hotel city choices by the clients")
plt.show()
```

```python
# Count of star ratings

val_counts = df["Etoiles"].value_counts()
print(val_counts, end = "\n\n")
plt.pie(val_counts, autopct = "%.2f%%", labels = val_counts.index)
plt.title('The chosen hotel star ratings')
plt.show()
```

```python
df['Nuits'].value_counts()
```

```python
#The impact of choosing the number of nights by the city of the hotel and the star rating.
sns.set_style("darkgrid")

plt.figure(figsize=(10, 6))
sns.barplot(x=df['Nuits'],y='Etoiles',hue='Ville',data=df)


plt.title('"The impact of choosing the number of nights by the city of the hotel and the star rating."')
plt.show()
```
